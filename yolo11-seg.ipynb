```python
!pip install ultralytics

from ultralytics import YOLO
import os

# Prevent duplicate library error
os.environ['KMP_DUPLICATE_LIB_OK'] = 'True'

# List of dataset YAML files (update paths accordingly)
dataset_yamls = [
    
    
    
    "/kaggle/input/secondd/data (4) (2).yaml",
    "/kaggle/input/thirddd/data (5) (2).yaml",
    "/kaggle/input/solar3-0-3/data (6).yaml",
    

]


# Training settings
epochs = 100  # Number of epochs
imgsz = 640   # Image size
batch_size = 16  # Batch size
device = 0  # GPU device ID (set to 'cpu' if no GPU available)

if __name__ == '__main__':
    for yaml_path in dataset_yamls:
        # Extract dataset name from the YAML file path for naming results
        dataset_name = os.path.basename(os.path.dirname(yaml_path))
        
        print(f"\nStarting training for dataset: {dataset_name}")
        print(f"Using YAML file: {yaml_path}")
        
        # Initialize YOLO model
        model = YOLO("yolo11n-seg.pt")
        
        # Train the model for the current dataset
        results = model.train(
            data=yaml_path,
            epochs=epochs,
            imgsz=imgsz,
            device=device,
            batch=batch_size,
            name=f"{dataset_name}_segmentation",  # Save results with dataset-specific name
        )
        
        print(f"Training complete for dataset: {dataset_name}")
```

```python
from ultralytics import YOLO
import cv2
import numpy as np

# Load trained YOLO models
yolo8 = YOLO('/kaggle/input/yoolo8/best (2) (1).pt')  # YOLOv8 model
yolo11 = YOLO('/kaggle/input/yolo11/best (3).pt')  # YOLOv11 trained model

# Load test image
image_path = '/kaggle/working/solar3.0-3/test/images/DJI_20240627100824_0008_T_JPG.rf.a99f38f53cefbceb94817810d2bb645a.jpg'
image = cv2.imread(image_path)

# Predict with YOLOv8
results8 = yolo8.predict(image)
masks8 = results8[0].masks.data.cpu().numpy()  # Binary masks
confidences8 = results8[0].probs.cpu().numpy()  # Confidence scores

# Predict with YOLOv11
results11 = yolo11.predict(image)
masks11 = results11[0].masks.data.cpu().numpy()
confidences11 = results11[0].probs.cpu().numpy()

# Simple ensemble - combine masks with confidence threshold
def ensemble_masks(mask1, conf1, mask2, conf2, threshold=0.5):
    combined_mask = np.zeros_like(mask1[0], dtype=np.uint8)
    
    # Go through each mask
    for i in range(mask1.shape[0]):
        combined_mask[(mask1[i] == 1) & (conf1[i] >= threshold)] = 1

    for j in range(mask2.shape[0]):
        combined_mask[(mask2[j] == 1) & (conf2[j] >= threshold)] = 1

    return combined_mask

# Combine the masks
combined_mask = ensemble_masks(masks8, confidences8, masks11, confidences11)

# Save the combined mask
cv2.imwrite('combined_mask.png', combined_mask * 255)

print("Combined mask saved successfully!")
```
